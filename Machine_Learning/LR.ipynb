```python
from category_encoders import OneHotEncoder
encoder = OneHotEncoder(use_cat_names = True)
df1 = encoder.fit_transform(df)
```

```python
from sklearn.model_selection import train_test_split
X = df1[df1.columns.difference(['Sweetness'])]
y = df1[['Sweetness']]
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
print(df1.shape)
```

```python
# target(Price)와 가장 correlated 된 features 를 k개 고르기.
## f_regresison, SelectKBest 불러오기.
from sklearn.feature_selection import f_regression, SelectKBest
## selctor 정의하기.
selector = SelectKBest(score_func=f_regression, k=10)
## 학습데이터에 fit_transform 
X_train_selected = selector.fit_transform(X_train, y_train)
## 테스트 데이터는 transform
X_test_selected = selector.transform(X_test)
X_train_selected.shape, X_test_selected.shape
```

```python
all_names = X_train.columns
## selector.get_support()
selected_mask = selector.get_support()
## 선택된 특성(변수)들
selected_names = all_names[selected_mask]
## 선택되지 않은 특성(변수)들
unselected_names = all_names[~selected_mask]
print('Selected names: ', selected_names)
print('Unselected names: ', unselected_names)
```

```python
features_10 = ['1rd_r', '2rd_r', '3rd_green', '3rd_rg', '4rd_r', '4rd_rg', '4th_green',
       '5rd_r', '6rd_r', '6th_green']
# 훈련데이터에 적용하기.
X_train[features_10].head()
from sklearn.metrics import r2_score
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
for alpha in [0, 0.001, 0.0025, 0.01, 0.05, 0.06, 0.07, 0.08, 0.09, 1.0, 1, 5, 10]:
        
    print(f'Ridge Regression, alpha={alpha}')
# Ridge 모델 만들기. 
    model = Ridge(alpha=alpha, normalize=True)  
    model.fit(X_train[features_10], y_train)
    y_pred = model.predict(X_test[features_10])
# MAE, R^2 for test 결과 확인하기.
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print(f'Test MAE: {mae:,.0f}')
    print(f'Test R2: {r2:,.3f}')
```
